```python
import pygame
#创建游戏窗口
import sys
import random

#素材位置
back_img = "img/background.png"
bird_1_img = "img/1.png"
bird_2_img = "img/2.png"
bird_dead_img = "img/dead.png"
pine_up_img = "img/top.png"
pine_down_img = "img/bottom.png"
wallX = 400  #管道x轴坐标
wallx_change = 5 #管道向左移动速度
pause = True
#鸟类
class Bird(object):
    def __init__(self):
        #鸟的判定矩形
        self.birdRect = pygame.Rect(65,50,50,50)
        #鸟的各种状态
        self.birdStatus = [pygame.image.load(bird_1_img),
                          pygame.image.load(bird_2_img),
                          pygame.image.load(bird_dead_img)]
        #状态表
        self.status = 0 #飞行状态
        self.birdX = 120 #x轴坐标 速度
        self.birdY = 350 #y轴坐标 高度
        self.jump = False  #掉落或者上升
        self.jumpSpeed = 10 #上升高度
        self.gravity = 5   #重力
        self.dead = False  #是否活着
        
        
    
    def birdUpdate(self):
        if self.jump:
             #跳跃动作
            self.jumpSpeed -= 1 #速度递减
            self.birdY -= self.jumpSpeed   #y轴坐标减小 小鸟上升
        else:
            #坠落动作
            self.gravity += 0.2
            self.birdY += self.gravity
        self.birdRect[1] = self.birdY #y轴位置变化 更改y轴

#管道类
class Pipeline(object):
    def __init__(self):
        self.wallx = wallX
        self.pineUp = pygame.image.load(pine_up_img)
        self.pineDown = pygame.image.load(pine_down_img)
        
        #管道移动
    def updatePipeline(self):
        self.wallx -= wallx_change #管道横向坐标递减5
        #穿越管道加分 并且重置
        if self.wallx < -80:
            global score
            score += 1
            self.wallx = 400
            

    
#创建地图
def createMap():
    screen.fill((255,255,255))
    screen.blit(background,(0,0))
    pygame.display.update()
    
    #显示管道
    screen.blit(Pipeline.pineUp,(Pipeline.wallx,-300))  #上管道坐标
    screen.blit(Pipeline.pineDown,(Pipeline.wallx,500))  #下管道坐标
    Pipeline.updatePipeline()#管道移动
    
    #显示小鸟
    if Bird.dead:   #小鸟去世
        Bird.status = 2
    elif Bird.jump:   #小鸟跳跃
        Bird.status = 1
    screen.blit(Bird.birdStatus[Bird.status],(Bird.birdX,Bird.birdY))    
    Bird.birdUpdate()   #小鸟移动
    
    #分数统计
    screen.blit(font.render('Score:'+str(score),-1,(255,255,255)),(100,50))
    pygame.display.update()
    
#状态检查
def checkDead():
    #管子矩形位置
    upRect = pygame.Rect(Pipeline.wallx,-300,
                         Pipeline.pineUp.get_width()-10,
                         Pipeline.pineUp.get_height())
    downRect = pygame.Rect(Pipeline.wallx,500,
                         Pipeline.pineDown.get_width()-10,
                         Pipeline.pineDown.get_height())
    #小鸟是否碰撞
    if upRect.colliderect(Bird.birdRect) or downRect.colliderect(Bird.birdRect):
        Bird.dead = True
    #是否出界
    if not 0< Bird.birdRect[1] < height:
        Bird.dead = True
        return True
    else:
        return False
    
#结束画面
def getResult():
    final_text1 = "GameOver"
    final_text2 = "Your score is:" + str(score)
    ft1_font = pygame.font.SysFont("Arial", 70)
    ft1_surf = font.render(final_text1, 1, (242, 3, 36))                             # 设置第一行文字颜色
    ft2_font = pygame.font.SysFont("Arial", 50)                                      # 设置第二行文字字体
    ft2_surf = font.render(final_text2, 1, (253, 177, 6))                            # 设置第二行文字颜色
    screen.blit(ft1_surf, [screen.get_width() / 2 - ft1_surf.get_width() / 2, 100])  # 设置第一行文字显示位置
    screen.blit(ft2_surf, [screen.get_width() / 2 - ft2_surf.get_width() / 2, 200])  # 设置第二行文字显示位置
    pygame.display.flip() 

#暂停
def text_objects(text, font):
    textSurface = font.render(text, True, (0, 0, 255))
    return textSurface, textSurface.get_rect()

def paused():
    largeText = pygame.font.SysFont('simsunnsimsun',80)
    TextSurf, TextRect = text_objects('琪琪的小鸟', largeText)
    TextRect.center = (200,200)
    screen.blit(TextSurf, TextRect)
     #按钮
    myimage=pygame.image.load(bird_1_img)
    screen.blit(myimage,[100,100])
    pygame.display.flip() 
    global pause
    while pause:
        for event in pygame.event.get():
            print(event)
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
            if event.type == pygame.MOUSEBUTTONDOWN:
                pause = False
        #pygame.display.update()
        clock.tick(15)
        
if __name__=='__main__':
    pygame.init()                          #初始化pygame
    pygame.font.init()
    font = pygame.font.SysFont("Arial",50)
    #窗口
    size = width,height = 400,650
    screen = pygame.display.set_mode(size) #显示窗口
    clock = pygame.time.Clock()
    #实例化
    Pipeline = Pipeline()                   #管道类实例化
    Bird = Bird()
    score = 0
    
    paused()
    while True:
        clock.tick(60)                       #
        #轮询问
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                sys.exit()
            if (event.type == pygame.KEYDOWN or event.type == pygame.MOUSEBUTTONDOWN) and not Bird.dead:
                Bird.jump = True  #跳跃
                Bird.gravity = 5   #重力
                Bird.jumpSpeed = 10 #跳跃速度
                
                
        #背景设置
        background = pygame.image.load(back_img)
        #小鸟状态
        if checkDead():
            getResult()
        else:
            createMap()
    pygame.quit()
    
```
